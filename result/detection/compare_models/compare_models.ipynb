```python
import os
from os.path import join as pj
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("ggplot")
%matplotlib inline
```

### --- compare accuracy distribution of size ---

```python
refinedet_not_pretrained = pd.read_csv('/home/tanida/workspace/Insect_project/figure/detection/RefineDet/crop_b2_2_4_8_16_32_im512_not_pretrain/accuracy_distribution_of_size.csv')
refinedet_pretrained = pd.read_csv('/home/tanida/workspace/Insect_project/figure/detection/RefineDet/crop_b2_2_4_8_16_32_im512/accuracy_distribution_of_size.csv')
faster_rcnn_not_pretrained = pd.read_csv('/home/tanida/workspace/Insect_project/figure/detection/Faster_RCNN/crop_b2_2_4_8_16_32_not_pretrain/accuracy_distribution_of_size.csv')
faster_rcnn_pretrained = pd.read_csv('/home/tanida/workspace/Insect_project/figure/detection/Faster_RCNN/crop_b2_2_4_8_16_32/accuracy_distribution_of_size.csv')
```

```python
all_df = refinedet_not_pretrained
all_df = all_df.rename(columns={'Accuracy': 'refinedet_not_pretrained'})
all_df = all_df.assign(refinedet_pretrained=refinedet_pretrained["Accuracy"])
all_df = all_df.assign(faster_rcnn_not_pretrained=faster_rcnn_not_pretrained["Accuracy"])
all_df = all_df.assign(faster_rcnn_pretrained=faster_rcnn_pretrained["Accuracy"])
all_df
```

```python
all_df.plot(x="Insect_size", y=["faster_rcnn_not_pretrained", "refinedet_not_pretrained", "faster_rcnn_pretrained", "refinedet_pretrained"], logx=True, legend=True, cmap="rainbow")
plt.ylabel("Precision")
plt.title("Precision distribution of size")
plt.savefig("/home/tanida/workspace/Insect_project/figure/detection/compare_precision_distrib_of_size.png", bbox_inches="tight")
```

### --- compare accuracy distribution of class ---

```python
refinedet_not_pretrained = pd.read_csv('/home/tanida/workspace/Insect_project/figure/detection/RefineDet/crop_b2_2_4_8_16_32_im512_not_pretrain/accuracy_distribution_of_class.csv')
refinedet_pretrained = pd.read_csv('/home/tanida/workspace/Insect_project/figure/detection/RefineDet/crop_b2_2_4_8_16_32_im512/accuracy_distribution_of_class.csv')
faster_rcnn_not_pretrained = pd.read_csv('/home/tanida/workspace/Insect_project/figure/detection/Faster_RCNN/crop_b2_2_4_8_16_32_not_pretrain/accuracy_distribution_of_class.csv')
faster_rcnn_pretrained = pd.read_csv('/home/tanida/workspace/Insect_project/figure/detection/Faster_RCNN/crop_b2_2_4_8_16_32/accuracy_distribution_of_class.csv')
```

```python
all_df = refinedet_not_pretrained
all_df = all_df.rename(columns={'Precision_per_class': 'refinedet_not_pretrained'})
all_df = all_df.assign(refinedet_pretrained=refinedet_pretrained["Precision_per_class"])
all_df = all_df.assign(faster_rcnn_not_pretrained=faster_rcnn_not_pretrained["Precision_per_class"])
all_df = all_df.assign(faster_rcnn_pretrained=faster_rcnn_pretrained["Precision_per_class"])
all_df
```

```python
all_df.plot(kind="bar", x="Name", y=["faster_rcnn_not_pretrained", "refinedet_not_pretrained", "faster_rcnn_pretrained", "refinedet_pretrained"], cmap="rainbow")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.ylabel("Precision")
plt.title("Precision per class")
plt.savefig("/home/tanida/workspace/Insect_project/figure/detection/compare_precision_per_class.png", bbox_inches="tight")
```
